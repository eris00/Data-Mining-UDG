```python
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt
```

```python
# Hardkodirani dataset
X = np.array([
    ['Mars', 'F/0/5', 'TRAPPIST-1e', 30, False],
    ['Venera', 'F/1/2', 'Proxima Centauri b', 45, True],
    ['Jupiter', 'M/4/7', 'Kepler-452b', 22, False],
    ['Mars', 'F/0/5', 'TRAPPIST-1e', 35, False],
    ['Saturn', 'M/2/1', 'TRAPPIST-1f', 28, True],
    ['Merkur', 'F/3/6', 'Proxima Centauri b', 52, True]
])
```

```python
y = [0, 0, 1, 1, 1, 0]  # Klasifikacioni target
```

```python
from sklearn.preprocessing import LabelEncoder

# Pretvaranje stringova u numeričke vrednosti
label_encoder = LabelEncoder()
for i in range(len(X[0])):
    if isinstance(X[0][i], str):
        X[:, i] = label_encoder.fit_transform(X[:, i])
```

```python
# Kreiranje i treniranje C4.5 stabla odlučivanja
clf = DecisionTreeClassifier(criterion='entropy')
clf.fit(X, y)
```

```python
# Vizualizacija stabla odlučivanja
plt.figure(figsize=(15, 5))
plot_tree(clf, filled=True, rounded=True, feature_names=['Maticna planeta', 'Kabina', 'Destinacija', 'Starost', 'VIP'], class_names=['Nije prevezen', 'Prevezen'])
plt.show()
```
