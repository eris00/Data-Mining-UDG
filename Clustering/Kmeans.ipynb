```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
# Unos dataset-a
X = np.array([[5.74, 165], [1.46, 71], [4.95,	140], [6.55,	150], [1.61,	63], [5.74,	125], [5.75,	158], [7.21,	215], [3.00, 77], [1.46,	62], [1.77,	70], [1.59,	78], [4.95,	140], [5.74, 180], [3.69,	90]])
```

```python
# Unos inicjalnih medoida
initial_centroids = np.array([[5.74, 165], [3.00, 77], [4.95,	140]])
```

```python
# Pokretanje K-means algoritma
K = len(initial_centroids)
kmeans = KMeans(n_clusters=K, init=initial_centroids, n_init=1)
labels = kmeans.fit_predict(X)
centroids = kmeans.cluster_centers_
```

```python
# Prikaz rezultata
colors = ['r', 'g', 'b']
plt.scatter(X[:, 0], X[:, 1], c=[colors[l] for l in labels], edgecolors='k')
plt.scatter(centroids[:, 0], centroids[:, 1], c='yellow', marker='x', s=100)
plt.title('K-means klasterovanje')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
# Ispisivanje klase za svaki podatak
for i, label in enumerate(labels):
    print(f"A{i+1}: Klasa {label+1}")
```
